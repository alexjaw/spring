## Dynamics of the movement

### We first set the initial conditions

```python
spring.constant = 2 # k
weight.mass = 10 # kg
weight.velocity = vector(0,0,0)
weight.acceleration = vector(0,0,0)
weight.force = vector(0,0,0)
```

The intial position is given by <code>spring.displacement=vector(0.35,0,0)</code>, displayed before.

### One step approximation

We start with a very rough approximation, by evolutionating the system from <code>spring.displacement=vector(0.35,0,0)</code> after a very long time increment: $\Delta t=2\ $seconds:

```python
dt=2
# Calculate the spring force using Hooke's Law
SpringForce = -spring.constant * spring.displacement

# The spring force acts on the weight
weight.force = SpringForce

# move the weight from the applied force using 
# Newton's 2nd law and the position equations
weight.acceleration = weight.force/weight.mass
weight.velocity =weight.velocity+weight.acceleration * dt
weight.pos =weight.pos+weight.velocity*dt + .5 * weight.acceleration * dt**2

print 'Average physical quantities after dt=',dt
print 'a=',weight.acceleration
print 'v=',weight.velocity
print 'x=',weight.pos

# calculate the new spring displacement
spring.displacement = weight.pos

# update the length of the spring
spring.axis = relaxedlength + spring.displacement
```

### Full numerical integration

We now reset to the initial condition again:

```python
spring.axis=relaxedlength
weight.pos=vector(0,0,0)
spring.displacement=vector(0.35,0,0)
spring.axis=spring.axis+spring.displacement
weight.pos=weight.pos+spring.displacement
```

The idea is to evolve step by step trough small time intervals according to the dynamic equation that governs of the system. The fast superposition of the small updated steps will create the illusion of movemement. Below is the full code of the simulation.

As now the $\Delta t$ will be very small, we need to keep only terms until order $\Delta t$ and safelly neglect terms of order $\Delta t^2$. Moreover, we can use directly the Principle of momentum
$$\mathbf{F}\Delta t=\Delta \mathbf{p}\,,$$
to obtain directly the velocity from the force:
$$\mathbf{v}=\mathbf{v}_0+\mathbf{F}/m\,.$$
In this way the acceleration does not appear explicitly in the dynamic equations (It is equivlent to assume tha the speed is constant along the small paths associated to each $\Delta t$) 

Finally, we can include the friction force proportional to the friction coefficient $\mu$, and recover the frictionless case just by settin $\mu=0$

The program models the motion of weight attached to a spring

```python
from __future__ import division
from visual import *

##########################################################################################
#
# INITIALIZE WINDOW & DECLARATIONS
#
##########################################################################################

scene.range = vector(1,1,1)
scene.center = vector(0,0,0)
scene.width = 800
scene.height = 600


##########################################################################################
#
# CREATE SPRING, WEIGHT & LABEL OBJECTS
#
##########################################################################################

relaxedlength = vector(.60,0,0) # length of spring when it isn't stretched or compressed
spring = helix(pos=(-.75,0,0),axis=relaxedlength, radius=.1,coils=8,thickness=.01,color=color.green)

weight = box(pos=(0,0,0),size=(.3,.3,.3),color=color.yellow)

frictionlessSurface = box(size=(2,.02,.5),pos=(0,-.16,0))
wall = box(size=(.04,.5,.3),pos=(-.77,.1,0),color=color.red)

##########################################################################################
#
# SET INITIAL CONDITIONS
#
########################################################################################## 
spring.constant = 2 # k
weight.mass = 10 # kg
weight.velocity = vector(0,0,0)
weight.acceleration = vector(0,0,0)
weight.force = vector(0,0,0)
```

__Set the initial position here__:

```python
#Set the label position
mylabel = label(pos=(0,.4,0))
#spring.displacement
xpos=0.35 #eval(raw_input('Enter initial displacement between -0.35 and 0.35\n'))
weight.pos=vector(xpos,0,0)
spring.displacement=weight.pos
spring.axis=relaxedlength+spring.displacement
# message to the user
message = "Intitial position:"
message += "\ndisplacement: %.2f" % spring.displacement.x 
mylabel.text = message

#raw_input('Hit <ENTER> to start simulation:')
```

#### Loop 

We will use <code>a+=b</code>  which is equivalent to <code>a=a+b </code>

```python
mu=0.0005
g=9.8 # m/s**2

dt = .01    # seconds
seconds = 0 # total time
steps=5000 #integration steps: final time= dt*steps

while seconds < dt*steps:
    rate(100)
    seconds += dt

    F_fr=vector(0,0,0)
    if weight.velocity.x>0:
        F_fr=vector(-mu*weight.mass*g,0,0)
    elif weight.velocity.x<0:
        F_fr=vector(mu*weight.mass*g,0,0)
    
    # Calculate the spring force using Hooke's Law
    SpringForce = -spring.constant * spring.displacement+F_fr

    # The spring force acts on the weight
    weight.force = SpringForce

    # move the weight from weight.pos using 
    # the momentum principle
    #weight.acceleration = weight.force/weight.mass
    weight.velocity += dt*weight.force/weight.mass # F*dt=dp
    weight.pos += weight.velocity*dt #+O(dt^2) -> + .5 * weight.acceleration * dt**2

    # calculate the new spring displacement
    spring.displacement = weight.pos

    # update the length of the spring
    spring.axis = relaxedlength + spring.displacement
    
    # message to the user
    message = "VPython Video Tutorial: Spring Force"
    message += "\ndisplacement: %.2f" % spring.displacement.x 
    message += "\nvelocity: %.3f" % weight.velocity.x
    message += "\nFriction: %.2f" % F_fr.x 
    message += "\nforce: % .1f" % weight.force.x
    mylabel.text = message
```
